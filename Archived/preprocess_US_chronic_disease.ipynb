```python
# Import depndencies 
import pandas as pd 
import numpy as np
from sqlalchemy import create_engine
from config import dbConnectionString
```

```python
# Create SQLEngine -- use "pip install psycopg2-binary" if ModuleError occurs
engine = create_engine(dbConnectionString)
```

```python
#load in data from sql database
Cardiovas_df = pd.read_sql_query('Select * FROM "USChronicDiseaseIndicatorsCDI" WHERE "Topic" like %(cvd)s',
            con=engine,
            params={"cvd":'Cardiovascular Disease'})
```

```python
Cardiovas_df.head()
```

```python
#drop unnecessary columns
Cardiovas_df.drop(columns = ['index','Question','Response','DataValueUnit','DataValueFootnoteSymbol','DatavalueFootnote','StratificationCategory2','Stratification2','StratificationCategory3','Stratification3','StratificationCategoryID2','StratificationID2','StratificationCategoryID3','StratificationID3'],
            axis=1, inplace=True)
Cardiovas_df.head()
```

```python
# drop the LocatioAbbr , DataValueAlt  column to make the date more clean
Cardiovas_df.drop('DataValueAlt', axis=1, inplace=True)
```

```python
#number of NaN values

Cardiovas_df["ResponseID"].isna().sum()
```

```python
# drop the QuestionID column to make the date more clean
Cardiovas_df.drop('QuestionID', axis=1, inplace=True)
# drop ResponseID  because all the values are NaN
Cardiovas_df.drop("ResponseID", axis=1 ,inplace=True)
```

```python
Cardiovas_df.head()
```

```python
# Replace Nan values with zeroes to make it easier for machine learning
#this dataset was eventually not used for ML 
Cardiovas_df['LocationDesc']=Cardiovas_df['LocationDesc'].fillna(0)
Cardiovas_df['DataSource']=Cardiovas_df['DataSource'].fillna(0)
Cardiovas_df['StratificationCategory1']=Cardiovas_df['StratificationCategory1'].fillna(0)
Cardiovas_df['Stratification1']=Cardiovas_df['Stratification1'].fillna(0)
Cardiovas_df['LocationID']=Cardiovas_df['LocationID'].fillna(0)
Cardiovas_df['TopicID']=Cardiovas_df['TopicID'].fillna(0)
Cardiovas_df['DataValueTypeID']=Cardiovas_df['DataValueTypeID'].fillna(0)
Cardiovas_df['DataValueTypeID']=Cardiovas_df['DataValueTypeID'].fillna(0)
Cardiovas_df['StratificationCategoryID1']=Cardiovas_df['StratificationCategoryID1'].fillna(0)
Cardiovas_df['StratificationID1']=Cardiovas_df['StratificationID1'].fillna(0)
Cardiovas_df['DataValue']=Cardiovas_df['DataValue'].fillna(0)
Cardiovas_df['LowConfidenceLimit']=Cardiovas_df['LowConfidenceLimit'].fillna(0)
Cardiovas_df['HighConfidenceLimit']=Cardiovas_df['HighConfidenceLimit'].fillna(0)
```

```python
Cardiovas_df.head()
```

```python
 # .str.replace to understand the structure of the column which parts will be removed  
    
Cardiovas_df['GeoLocation'].str.replace(r"([POINT])", r" \1").str.split(expand=True)
```

```python
# Removing POINT and the parentheses () from the column
Cardiovas_df['GeoLocation'] = Cardiovas_df['GeoLocation'].str.replace('[POINT]', '')
Cardiovas_df['GeoLocation'] = Cardiovas_df['GeoLocation'].str.replace('[()]', '')
```

```python
temp_df = Cardiovas_df
temp_df.head()
```

```python
Cardiovas_df[['longitude','latitude']] = Cardiovas_df['GeoLocation'].str.lstrip().str.rstrip().str.split(' ',1,expand=True)
```

```python
Cardiovas_df[Cardiovas_df['longitude'].isna()]
Cardiovas_df.dropna(subset=['longitude','latitude'],axis=0, inplace=True)
```

```python
Cardiovas_df[["longitude", "latitude"]] = Cardiovas_df[["longitude", "latitude"]].apply(pd.to_numeric)
```
